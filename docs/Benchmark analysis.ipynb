```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data from the table
data = {
    "LoC": ["1 - 25", "26 - 50", "51-75", "75+"],
    "One-shot prompting": [35.9375, 58.53658537, 71.42857143, 66.66666667],
    "Few-shots prompting": [38.88888889, 62.06896552, 44.44444444, 100],
    "Chain of Thoughts": [24.07407407, 41.46341463, 50, 83.33333333]
}

df = pd.DataFrame(data)

# Converting failure rates to success rates
df['One-shot prompting'] = 100 - df['One-shot prompting']
df['Few-shots prompting'] = 100 - df['Few-shots prompting']
df['Chain of Thoughts'] = 100 - df['Chain of Thoughts']

# Sample counts
samples = [59, 43, 11, 9]

# Plotting the bar graph
labels = df['LoC']
one_shot = df['One-shot prompting'] / 100
few_shots = df['Few-shots prompting'] / 100
chain_of_thoughts = df['Chain of Thoughts'] / 100

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, one_shot, width, label='One-shot prompting', color='blue')
rects2 = ax.bar(x, few_shots, width, label='Few-shots prompting', color='orange')
rects3 = ax.bar(x + width, chain_of_thoughts, width, label='Chain of Thoughts', color='green')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Success Rate') 
ax.set_xticks(x)
ax.set_xticklabels([f'{label}\n [n={sample}]' for label, sample in zip(labels, samples)], rotation=45)
ax.set_ylim(0, 1)  # Set y-axis range from 0 to 1
ax.legend()

fig.tight_layout()

plt.show()
 
    
```
